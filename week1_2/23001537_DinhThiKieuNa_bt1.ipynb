4. Các toán tử

```python
x = 5
y = 10

print(x + y)
print(x - y)
print(x * y)
print(x / y)
print(x // y)
print(x % y)
print(x ** y)

print(x == y)
print(x != y)
print(x > y)
print(x < y)
print(x >= y)
print(x <= y)
```

7. Dictionary

```python
my_dictionary = {"Na": "KHTN", "Hoa" : "DHCN", "An": "DHVN"}
print(my_dictionary["An"])
hoa = my_dictionary.get("Hoa")
oanh = my_dictionary.get("Oanh", "Unknown")
print(hoa)
print(oanh)
```

```python
print(my_dictionary.items())
print(my_dictionary.keys())
print(my_dictionary.values())
```
