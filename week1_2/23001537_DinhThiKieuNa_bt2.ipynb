7. Function

```python
def avg_grade(math, english, literal):
    return (math  * 3 + literal * 2 + english) / 6
math = int(input("Enter your math score: "))
english = int(input("Enter your english score: "))
literal = int(input("Enter your literal score: "))
print(f"Your average score {avg_grade(math, english, literal)}")
```

8. Class

8.1. Kế thừa

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def introduce(self):
        return f"My name is {self.name} and I am {self.age} years old."
    
alice = Person("Alice", 30)
print(alice.introduce())
```

```python
class Employee(Person):
    def __init__(self, name, age, employee_id):
        super().__init__(name, age)
        self.employee_id = employee_id
    def introduce(self):
        return  f"My name is {self.name},I am {self.age} years old, and my employee id is {self.employee_id}"
join = Employee("John", 35, "EMP123")
print(join.introduce())
```

```python
class Student(Person):
    def __init__(self, name, age, student_id):
        super().__init__(name, age)
        self.student_id = student_id
    def introduce(self):
        return  f"My name is {self.name},I am {self.age} years old, and my student id is {self.student_id}"
    def hello(self):
        return f"My name is {self.name}, I just want to say hello to you guys"
join =Student("John", 35, "ST123")
print(join.introduce())
print(join.hello())
```

8.2. Kế thừa đa hình

```python
class Animal:
    def make_sound(self):
        pass

class Dog(Animal):
    def make_sound(self):
        return "Growww growwwww"
    
class Cat(Animal):
    def make_sound(self):
        return "Meowwwwww"
    
def animal_sounds(animal):
    return animal.make_sound()
    
dog = Dog()
cat = Cat()

print(animal_sounds(dog))
print(animal_sounds(cat))
```
